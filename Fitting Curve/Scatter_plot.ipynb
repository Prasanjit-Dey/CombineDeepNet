```python
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np
from numpy import savetxt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
predicted=read_csv("Predict_Aotizh.csv")
predicted.values
```

```python
predicted = predicted.values.ravel()
predicted.shape
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
actual=read_csv("Test_Aotizh.csv")
actual.values
```

```python
actual = actual.values.ravel()
actual.shape
```

```python

rng = np.random.RandomState(0)
x = rng.randn(180)
y = rng.randn(180)
colors = rng.rand(180)
sizes = sizes = 180
plt.scatter(actual, predicted, c=colors, s=sizes, alpha=0.7,
            cmap='Greens')

# add a line of best fit
#fit = np.polyfit(actual, predicted, 1)
fit = np.polyfit(actual, predicted, 1)
fit_fn = np.poly1d(fit)
plt.plot(actual, fit_fn(actual), '--k', linewidth=1.5, c= "Red")

# add labels and title to the plot
plt.xlabel('Actual PM2.5 cocentration')
plt.ylabel('Predicted PM2.5 cocentration')
#plt.title('Scatter Plot of Actual vs Predicted Values')

# display the plot
#plt.colorbar()
plt.savefig(r'./Scatter_Aotizhongxin.jpg', dpi=400)
plt.show()
```
